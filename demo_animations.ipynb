```python
from ising_model import ClassicIsing
import electron
from grid import Grid, HoleGrid, Mobius, Cylinder, Torus
import numpy as np
from IPython.display import Video
from visualizations import animate_ising

# Before running the code, make sure to reset the kernel to clear any old imports.
```

## Regular Grid

```python
newGrid = Grid(n_x=10, n_y=10, gridPointObject=electron.ClassicElectron, random_seed=12)
isingModel = ClassicIsing(newGrid, temperature=1, ferromagnetivity=.1,Mf_External=0,)

isingModel.runSimulation(1000)


ani = animate_ising(newGrid, isingModel.grid.grid_history, interval=50, title="Regular Grid Evolution")

# save animation as mp4 file and embed it
ani.save('ising_evolution.mp4', writer='ffmpeg', fps=20)
Video('ising_evolution.mp4', embed=True)
```

##Hole Grid

```python
holeGrid = HoleGrid(n_x=10, n_y=10, gridPointObject=electron.ClassicElectron, random_seed=2, hole_grid=np.ones((3,3)) ,c_y=2, c_x=2)
isingModel = ClassicIsing(holeGrid, temperature=1, ferromagnetivity=1,Mf_External=0,)

isingModel.runSimulation(10)

ani = animate_ising(holeGrid, isingModel.grid.grid_history, interval=100, title="Hole Grid Evolution")

ani.save('ising_evolution.mp4', writer='ffmpeg', fps=10)
Video('ising_evolution.mp4', embed=True)
```
